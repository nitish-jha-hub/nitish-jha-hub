```python
a=[1,2,3]
b=a
print(a is b) # strictly equality operator as  === in js

# Example with == (checks for equality of value)
a = [1, 2, 3]
b = [1, 2, 3]
print(f"a == b: {a == b}")

c = a
print(f"a == c: {a == c}")

# Example with is (checks if two variables refer to the same object)
a = [1, 2, 3]
b = [1, 2, 3]
print(f"a is b: {a is b}")

c = a
print(f"a is c: {a is c}")
```

As you can see from the output:

- `a == b` is `True` because the values of the lists `a` and `b` are the same, even though they are different objects in memory.
- `a == c` is `True` because `c` is assigned the same list as `a`, so they have the same value and refer to the same object.
- `a is b` is `False` because `a` and `b` are distinct list objects in memory, even though they have the same contents.
- `a is c` is `True` because `c = a` makes both variables refer to the exact same list object in memory.

```python
my_dict = {"name": "Alice", "age": 30}
print(f"Initial dictionary: {my_dict}")

# 1. Adding a new key-value pair using direct assignment
my_dict["city"] = "New York"

# 2. Updating an existing key-value pair using direct assignment
my_dict["age"] = 31

# 3. Adding/updating multiple key-value pairs using update()
my_dict.update({"occupation": "Engineer", "city": "San Francisco"})

# 4. Adding an element to a list that is a dictionary value
my_dict["hobbies"] = ["reading", "hiking"]

my_dict["hobbies"].append("cooking")
```

Request package to call api


```python
import requests

# Example: Search books with keyword "JavaScript"
url = "https://www.googleapis.com/books/v1/volumes?q=javascript"

response = requests.get(url)

if response.status_code == 200:
    data = response.json()
    print("Total Items:", data.get("totalItems"))
    if "items" in data:
        print("First Book Title:", data["items"][0]["volumeInfo"]["title"])

else:
    print("Failed with status:", response.status_code)
```
